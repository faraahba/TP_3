## *Question 2*

```python
#importation des bibliotheques
import numpy as np
import matplotlib.pyplot as plt

def compute_error_for_line_given_point(b,w,points):
    totalerror=0
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        totalerror+=(y-(w*x+b))**2
    return totalerror/(2*float(len(points)))
```

## *Question 3*

```python
#fonction de descente de gradient pour mettre à jour les paramètres (b, w) d'un modèle linéaire.
def step_gradent(b_current,w_current,points,learnrate):
    b_gradient=0
    w_gradient=0
    N=float(len(points))

    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        #Calcul des gradients pour le terme de biais et le poids
        b_gradient+=(1/N)*((w_current*x+b_current)-y)#
        w_gradient+=(1/N)*x*((w_current*x+b_current)-y)
    #Mise à jour des paramètres
    new_b=b_current-(learnrate*b_gradient)
    new_w=w_current-(learnrate*w_gradient)
    return [new_b,new_w]
```

## *Question 4*

```python
# Exécute l'algorithme de descente de gradient pour ajuster les paramètres (b, w) d'un modèle linéaire.
def gradient_descent_runner(points,starting_b,starting_w,learning_rate,num_iteration):
    b=starting_b
    w=starting_w
    cost_history = np.zeros(num_iteration)
    for i in range(num_iteration):
          # Effectuer une étape de descente de gradient
        b,w=step_gradent(b,w,np.array(points),learning_rate)
        print("Iteration Numero :",i)
        print("Fonction du cout = ", compute_error_for_line_given_point(b,w,np.array(points)))
            # Calculer et enregistrer le coût de l'itération actuelle
        cost_history[i] = compute_error_for_line_given_point(b,w,np.array(points))
        print("w=", w)
        print("b=", b)
    # Afficher l'évolution du coût au fil des itérations
    plt.plot(range(num_iteration), cost_history)
    plt.show()
    # Retourner les valeurs finales du terme de biais et du poids
    return [b,w]
```

## *Question 5*

```python
# Hyperparameter : learning_rate = 0.0001 et nombre_itération = 100 :
def run():

    # Initialize one hundred points
    x=np.linspace(1,10,100)
    y=np.linspace(1,10,100)

    array=[]

    for i in np.arange(100):
        for j in np.arange(100):
            p=[x[i],y[i]]
            array.append(p)

 # Définir les hyperparamètres
    learning_rate=0.0001
    init_b=0
    init_w=0
    num_iteration=100
# Exécuter la descente de gradient
    [b,w]=gradient_descent_runner(array,init_b,init_w,learning_rate,num_iteration)
# Tracer les points d'origine et la ligne de régression
    plt.figure()
    plt.scatter(x, y)
    y2=b+w*x
    plt.scatter(x,y2)
# Affichage des valeurs finales de b et w    
    print("last_b=", b)
    print("last_w=", w)

if __name__=="__main__":
    run()
```

## *Question 1*

```python
# Importation des bibliothèques
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
```

## *Question 2*

```python
# Génération de données synthétiques pour la régression :
np.random.seed(0)
X, y = make_regression(n_samples=100, n_features=1, noise=10)
```

## *Question 3*

```python
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Visualisation des données')
plt.show()
```

## *Question 4*

```python
# Création d'un modèle de régression linéaire utilisant la descente de gradient stochastique (SGD)
model = SGDRegressor(max_iter=100, alpha=0.0001)
```

## *Question 5*

```python
# Entraînement du modèle de régression linéaire sur les données d'entraînement
model.fit(X, y)
```

## *Question 6*

```python
# Calcul du coefficient de détermination R² pour évaluer la qualité de l'ajustement du modèle aux données
print("Coefficient de détermination (R^2) :", model.score(X,y))
     
```

## *Question 7*

```python
# Tracer les données d'entraînement et la ligne de régression prédite par le modèle
predictions = model.predict(X)
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red', lw=3)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Prédictions du modèle')
plt.show()
```

## *Question 8*

```python
# Création d'un nouveau modèle de régression linéaire utilisant la descente de gradient stochastique (SGD), avec des hyperparamètres modifiés
model_high_iterations = SGDRegressor(max_iter=1000, alpha=0.001)
model_high_iterations.fit(X, y)
```

## *Question 9*

```python
# Calcul du coefficient de détermination R² pour évaluer la qualité de l'ajustement du nouveau modèle aux données
# Tracer à nouveau les données d'entraînement et la nouvelle ligne de régression prédite par le modèle
accuracy_high_iterations = model_high_iterations.score(X, y)
print("Coefficient de détermination (R^2) :", accuracy_high_iterations)
predictions_high_iterations = model_high_iterations.predict(X)
plt.scatter(X, y)
plt.plot(X, predictions_high_iterations, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Prédictions du modèle avec 1000 itérations')
plt.show()
```
